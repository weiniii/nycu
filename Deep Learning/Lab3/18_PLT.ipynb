```python
import matplotlib.pyplot as plt 
import pylab as pl
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import torch
import numpy
```

```python
NOpre_train = torch.load('80/18nopre/Resnet18_nopretrain.pt')
NOpre_test = torch.load('80/18nopre/Resnet18_nopretest.pt')
Pre_train = torch.load('80/18pre_80/Resnet18_pretrain.pt')
Pre_test = torch.load('80/18pre_80/Resnet18_pretest.pt')
```

```python
Pre_train
```

```python
Pre_test
```

```python
x = []
for i in range(1,11):
    x.append(i)
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1,1,1)
ax1.set_title('ResNet18',fontsize=25)
plt.xlabel('Epoch',fontsize=20)
plt.ylabel('Accuracy(%)',fontsize=20)
p2 = pl.plot(x,np.array(torch.tensor(NOpre_train,device='cpu')),'g-',label=u' Train(w/o pretraning)', linewidth=1)
pl.legend()
p3 = pl.plot(x,np.array(torch.tensor(Pre_train,device='cpu')),'r-',label=u' Train(with pretraning)', marker='o', linewidth=1)
pl.legend()
p4 = pl.plot(x,np.array(torch.tensor(NOpre_test,device='cpu')),'b-',label=u' Test(w/o pretraning)', linewidth=1)
pl.legend()
p5 = pl.plot(x,np.array(torch.tensor(Pre_test,device='cpu')),'c-',label=u' Test(with pretraning)', marker='o', linewidth=1)
pl.legend()
plt.grid()
```
